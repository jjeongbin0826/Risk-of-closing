```python
mapping = pd.read_excel('/content/drive/Shareddrives/서울시 빅데이터(2022-2)/논문/0. 원본데이터/임대료/임대료매핑.xlsx', sheet_name=1)
Acode = pd.read_csv('/content/drive/Shareddrives/서울시 빅데이터(2022-2)/논문/0. 원본데이터/임대료/행정동코드.csv')

print(mapping.head())
print(Acode.head())
```

```python
Acode = Acode[['행정구역', 'ADSTRD_CD']]
Acode = Acode.drop_duplicates()
Acode
```

```python
trade = pd.merge(mapping, Acode, on='행정구역', how='left')
trade
```

```python
trade = trade.drop('행정구역', axis=1)
trade = trade.dropna(how='all')
trade.info()
```

```python
trade.dropna(subset=['ADSTRD_CD'], inplace=True)
trade
```

```python
year_quarter = trade.columns.str.extract(r'(\d+)년 (\d)분기')
year = year_quarter[0].astype(float).fillna(0).astype(int)
```

```python
trade_T = []

for index, row in trade.iterrows():

    year_quarter = trade.columns.str.extract(r'(\d+)년 (\d)분기')
    year = year_quarter[0].astype(float).fillna(0).astype(int)
    quarter = year_quarter[1].astype(float).fillna(0).astype(int)

    adstrd_cd = int(row['ADSTRD_CD'])

    rent = row.values[:-1]

    for i in range(min(len(year), len(quarter), len(rent))):
        trade_T.append((year[i], quarter[i], adstrd_cd, rent[i]))

trade_T = pd.DataFrame(trade_T, columns=['년도', '분기', '행정동코드', '임대료'])
trade_T
```

```python
trade_T['행정동코드']=trade_T['행정동코드']//100
trade_T
```

```python
rental['환산임대료평균'] = rental.apply(lambda row: trade_T.loc[(trade_T['년도'] == row['년도']) & (trade_T['분기'] == row['분기']) & (trade_T['행정동코드'] == row['행정동코드']), '임대료'].values[0]
                                 if (row['환산임대료평균'] == 0) and (not trade_T.loc[(trade_T['년도'] == row['년도']) & (trade_T['분기'] == row['분기']) & (trade_T['행정동코드'] == row['행정동코드']), '임대료'].empty) else row['환산임대료평균'], axis=1)
rental
```

```python
rental = rental.loc[(rental['환산임대료평균'] != 0) & (~rental['환산임대료평균'].isna())]
rental.info()
```

```python
rental = rental[['년도', '분기', '행정동코드', '환산임대료평균', '임대건수']]
rental = rental.sort_values(['년도', '분기', '행정동코드'])
rental = rental.reset_index(drop=True)
rental = rental.drop_duplicates()
rental
```

```python
income=pd.read_csv('/content/drive/Shareddrives/서울시 빅데이터(2022-2)/논문/1. 전처리/income.csv')
income_temp=income[['년도', '분기', '행정동코드']]
income_temp=income_temp.drop_duplicates()

temp=pd.merge(income_temp, rental, on=['년도', '분기', '행정동코드'], how='outer')
temp.info()
```

```python
temp.sort_values(by=['년도', '분기'], inplace=True)
temp['순서'] = temp.groupby(['년도', '분기']).ngroup() + 1

temp
```

```python
columns_to_fill = ['환산임대료평균', '임대건수']

for column in columns_to_fill:
    for index, row in temp[temp[column].isnull()].iterrows():
        candidate_rows = temp[
            (temp['행정동코드'] == row['행정동코드']) &
            ((temp['순서'] == row['순서'] - 1) | (temp['순서'] == row['순서'] + 1))
        ]

        if not candidate_rows.empty:
            avg_value = candidate_rows[column].mean()
            temp.at[index, column] = avg_value

temp.info()
```

```python
temp['행정동코드']=temp['행정동코드'].astype('int64')
temp=pd.merge(temp,code,on=['행정동코드'], how='left')
temp.info()
```

```python
fill_mean_func = lambda g: g.fillna(g.mean())
temp=temp.groupby(['년도','분기','시군구코드']).apply(fill_mean_func)
temp.info()
```

```python
rental=pd.merge(income, temp, on=['년도', '분기', '행정동코드'], how='left')
rental
```

```python
rental=rental.dropna()
rental=rental.drop(['시군구코드', '순서'],axis=1)
rental
```

```python
rental.to_csv('./rental.csv', index=False)
```

```python
temp.sort_values(by=['년도', '분기'], inplace=True)
temp['순서'] = temp.groupby(['년도', '분기']).ngroup() + 1

temp
```

```python
columns_to_fill = ['평균_생활인구_수', '월요일_생활인구_수', '화요일_생활인구_수', '수요일_생활인구_수',
                   '목요일_생활인구_수', '금요일_생활인구_수', '토요일_생활인구_수', '일요일_생활인구_수']

for column in columns_to_fill:
    for index, row in temp[temp[column].isnull()].iterrows():
        candidate_rows = temp[
            (temp['행정동코드'] == row['행정동코드']) &
            ((temp['순서'] == row['순서'] - 1) | (temp['순서'] == row['순서'] + 1))
        ]

        if not candidate_rows.empty:
            avg_value = candidate_rows[column].mean()
            temp.at[index, column] = avg_value

temp.info()
```

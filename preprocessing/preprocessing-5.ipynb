```python
mydata=pd.read_csv('/content/drive/Shareddrives/서울시 빅데이터(2022-2)/논문/1. 전처리/mydata.csv')
mydata=mydata[(mydata['개업년월']>=202010)&(mydata['개업년월']<=202206)]

mydata['년도']=mydata['개업년월']//100
mydata['월']=mydata['개업년월']%100

conditions=[(mydata['월']>=1)&(mydata['월']<=3),
           (mydata['월']>=4)&(mydata['월']<=6),
           (mydata['월']>=7)&(mydata['월']<=9),
           (mydata['월']>=10)&(mydata['월']<=12)]
values=[1,2,3,4]
mydata['분기']=np.select(conditions,values)

mydata
```

```python
frame=mydata[['년도', '분기', '행정동코드', '표준산업분류코드']]
frame=frame.drop_duplicates()
frame
```

```python
frame['년도_1'] = frame.apply(lambda row: row['년도'] - 1 if row['분기'] == 1 else row['년도'], axis=1)
frame['분기_1'] = frame.apply(lambda row: 4 if row['분기'] == 1 else row['분기'] - 1, axis=1)

frame['년도_2'] = frame.apply(lambda row: row['년도_1'] - 1 if row['분기_1'] == 1 else row['년도_1'], axis=1)
frame['분기_2'] = frame.apply(lambda row: 4 if row['분기_1'] == 1 else row['분기_1'] - 1, axis=1)

frame['년도_3'] = frame.apply(lambda row: row['년도_2'] - 1 if row['분기_2'] == 1 else row['년도_2'], axis=1)
frame['분기_3'] = frame.apply(lambda row: 4 if row['분기_2'] == 1 else row['분기_2'] - 1, axis=1)

frame['년도_1'] = frame['년도_1'].astype('int64')
frame['분기_1'] = frame['분기_1'].astype('int64')

frame['년도_2'] = frame['년도_2'].astype('int64')
frame['분기_2'] = frame['분기_2'].astype('int64')

frame['년도_3'] = frame['년도_3'].astype('int64')
frame['분기_3'] = frame['분기_3'].astype('int64')

frame
```

```python
frame = frame.drop(['년도_1', '분기_1'],axis=1)
frame
```

```python
people_2=pd.read_csv('/content/drive/Shareddrives/서울시 빅데이터(2022-2)/논문/1. 전처리/people.csv')
people_2.columns = [f'{col}_2' for col in people_2.columns]
people_2=people_2.rename(columns={'행정동코드_2':'행정동코드', '표준산업분류코드_2':'표준산업분류코드'})

data=pd.merge(frame, people_2, on=['년도_2', '분기_2', '행정동코드', '표준산업분류코드'], how='left')
data.info()
```

```python
people_3=pd.read_csv('/content/drive/Shareddrives/서울시 빅데이터(2022-2)/논문/1. 전처리/people.csv')
people_3.columns = [f'{col}_3' for col in people_3.columns]
people_3=people_3.rename(columns={'행정동코드_3':'행정동코드', '표준산업분류코드_3':'표준산업분류코드'})

data=pd.merge(data, people_3, on=['년도_3', '분기_3', '행정동코드', '표준산업분류코드'], how='left')
data.info()
```

```python
data
```

```python
data=data.dropna()
data=data.drop(columns = ['년도_2', '분기_2', '년도_3', '분기_3'])
data=data.drop_duplicates()
data=data.sort_values(['년도', '분기', '행정동코드', '표준산업분류코드'])
data=data.reset_index(drop=True)
data
```

```python
index[['년도']] = index[['date']]//100
index[['월']] = index[['date']]%100
index
```

```python
conditions=[(index['월']>=1)&(index['월']<=3),
           (index['월']>=4)&(index['월']<=6),
           (index['월']>=7)&(index['월']<=9),
           (index['월']>=10)&(index['월']<=12)]
values=[1,2,3,4]
index['분기']=np.select(conditions,values)
index
```

```python
index=index.drop(['date', '월'],axis=1)
index=index.groupby(['년도','분기','행정동_코드']).agg({'grade':np.mean,'result':np.mean,'sales':np.mean, 'infra':np.mean, 'franch':np.mean, 'pop':np.mean, 'fin':np.mean}).reset_index()
index
```

```python
index=index.rename(columns={'행정동_코드':'행정동코드'})
index=index[['년도', '분기', '행정동코드', 'grade', 'result', 'sales', 'infra', 'franch', 'pop', 'fin']]
index=index.sort_values(['년도', '분기', '행정동코드'])
index=index.reset_index(drop=True)
index=index.drop_duplicates()
index
```

```python
frame=pd.read_csv('/content/drive/Shareddrives/서울시 빅데이터(2022-2)/논문/1. 전처리/frame.csv')
frame_temp=frame[['년도', '분기', '행정동코드']]
frame_temp=frame_temp.drop_duplicates()

temp=pd.merge(frame_temp, index, on=['년도', '분기', '행정동코드'], how='outer')
temp
```

```python
temp.sort_values(by=['년도', '분기'], inplace=True)
temp['순서'] = temp.groupby(['년도', '분기']).ngroup() + 1

temp
```

```python
columns_to_fill = ['grade', 'result', 'sales', 'infra', 'franch', 'pop', 'fin']

for column in columns_to_fill:
    for index, row in temp[temp[column].isnull()].iterrows():
        candidate_rows = temp[
            (temp['행정동코드'] == row['행정동코드']) &
            ((temp['순서'] == row['순서'] - 1) | (temp['순서'] == row['순서'] + 1))
        ]

        if not candidate_rows.empty:
            avg_value = candidate_rows[column].mean()
            temp.at[index, column] = avg_value

temp.info()
```

```python
temp['행정동코드']=temp['행정동코드'].astype('int64')
temp=pd.merge(temp,code,on=['행정동코드'], how='left')
temp.info()
```

```python
fill_mean_func = lambda g: g.fillna(g.mean())
temp=temp.groupby(['년도','분기','시군구코드']).apply(fill_mean_func)
temp.info()
```

```python
index=pd.merge(frame, temp, on=['년도', '분기', '행정동코드'], how='left')
index
```

```python
index=index.dropna()
index=index.drop(['시군구코드', '순서'],axis=1)
index
```

```python
index.to_csv('./index.csv', index=False)
```

```python
temp.sort_values(by=['년도', '분기'], inplace=True)
temp['순서'] = temp.groupby(['년도', '분기']).ngroup() + 1

temp
```

```python
columns_to_fill = ['분기당_매출_금액', '분기당_매출_건수', '월요일_매출_금액', '화요일_매출_금액', '수요일_매출_금액', '목요일_매출_금액', '금요일_매출_금액',
                   '토요일_매출_금액', '일요일_매출_금액', '월요일_매출_건수', '화요일_매출_건수', '수요일_매출_건수', '목요일_매출_건수', '금요일_매출_건수',
                   '토요일_매출_건수', '일요일_매출_건수', '점포수']

for column in columns_to_fill:
    for index, row in temp[temp[column].isnull()].iterrows():
        candidate_rows = temp[
            (temp['행정동코드'] == row['행정동코드']) &
            (temp['표준산업분류코드'] == row['표준산업분류코드']) &
            ((temp['순서'] == row['순서'] - 1) | (temp['순서'] == row['순서'] + 1))
        ]

        if not candidate_rows.empty:
            avg_value = candidate_rows[column].mean()
            temp.at[index, column] = avg_value

temp.info()
```

```python
temp['행정동코드']=temp['행정동코드'].astype('int64')
temp=pd.merge(temp,code,on=['행정동코드'], how='left')
temp.info()
```

```python
fill_mean_func = lambda g: g.fillna(g.mean())
temp=temp.groupby(['년도','분기','표준산업분류코드','시군구코드']).apply(fill_mean_func)
temp.info()
```

# mydata 전처리

```python
mydata=pd.read_csv("/content/drive/Shareddrives/서울시 빅데이터(2022-2)/논문/1. 전처리/mydata.csv")
mydata=mydata[['점포명', '개업년월', '폐업년월', '표준산업분류코드', '행정동코드', '프랜차이즈여부']]
mydata
```

```python
mydata_Y=mydata[~(mydata['폐업년월']==0)]
mydata_N=mydata[mydata['폐업년월']==0]

print(mydata_Y.head())
print(mydata_N.head())
```

```python
mydata_N['폐업여부']='N'
mydata_N
```

```python
mydata_Y['개업년'] = mydata_Y['개업년월'] // 100
mydata_Y['개업월'] = mydata_Y['개업년월'] % 100
mydata_Y['폐업년'] = mydata_Y['폐업년월'] // 100
mydata_Y['폐업월'] = mydata_Y['폐업년월'] % 100

mydata_Y['폐업여부'] = 'N'

mydata_Y.loc[
    ((mydata_Y['폐업년'] - mydata_Y['개업년']) * 12 + (mydata_Y['폐업월'] - mydata_Y['개업월']) <= 6)
    & (~mydata_Y['폐업년월'].isna()), '폐업여부'
] = 'Y'

mydata_Y=mydata_Y.drop(columns = ['개업년', '개업월', '폐업년', '폐업월'])

mydata_Y
```

```python
mydata=pd.concat([mydata_Y, mydata_N], axis = 0)
mydata=mydata[(mydata['개업년월']>=202000) & (mydata['개업년월']<=202300)]
mydata=mydata.sort_values(by=['개업년월', '행정동코드', '표준산업분류코드'])
mydata=mydata.reset_index(drop=True)
mydata
```

```python
mydata['년도']=mydata['개업년월']//100
mydata['월']=mydata['개업년월']%100

conditions=[(mydata['월']>=1)&(mydata['월']<=3),
           (mydata['월']>=4)&(mydata['월']<=6),
           (mydata['월']>=7)&(mydata['월']<=9),
           (mydata['월']>=10)&(mydata['월']<=12)]
values=[1,2,3,4]
mydata['분기']=np.select(conditions,values)

mydata=mydata.drop(columns = ['개업년월', '폐업년월', '월'])
mydata=mydata[['점포명', '년도', '분기', '행정동코드', '표준산업분류코드', '프랜차이즈여부', '폐업여부']]
mydata
```

# mydata+pred

## 1) x -> y

```python
data_pred=pd.read_csv("/content/drive/Shareddrives/서울시 빅데이터(2022-2)/논문/2. 상권 활성화 지수 예측 모델/3) 예측 데이터/x_forecast_LR.csv")
data_x=data_pred.iloc[:, 0:88]
data_pred=data_pred[['년도', '분기', '행정동코드', '표준산업분류코드', 'grade_1_pre', 'result_1_pre', 'sales_1_pre', 'infra_1_pre', 'franch_1_pre', 'pop_1_pre', 'fin_1_pre']]
data_pred.columns=['년도', '분기', '행정동코드', '표준산업분류코드', 'grade_pre', 'result_pre', 'sales_pre', 'infra_pre', 'franch_pre', 'pop_pre', 'fin_pre']

print(data_x.columns)
print(data_pred.columns)
```

```python
data_total=pd.merge(mydata, data_x, on=['년도', '분기', '행정동코드', '표준산업분류코드'], how='left')
data_total=data_total.dropna()
data_total
```

```python
data_total_x=data_total.drop('폐업여부', axis=1)
data_total_y=data_total[['점포명', '년도', '분기', '행정동코드', '표준산업분류코드', '폐업여부']]

print(data_total_x.head())
print(data_total_y.head())
```

```python
data_total_x.to_csv('./data_total_x.csv', index=False)
data_total_y.to_csv('./data_total_y.csv', index=False)
```

## 2) i_hat -> y

```python
data_PredIndex=pd.merge(mydata, data_pred, on=['년도', '분기', '행정동코드', '표준산업분류코드'], how='left')
data_PredIndex=data_PredIndex.dropna()
data_PredIndex
```

```python
data_PredIndex_x=data_PredIndex.drop('폐업여부', axis=1)
data_PredIndex_y=data_PredIndex[['점포명', '년도', '분기', '행정동코드', '표준산업분류코드', '폐업여부']]

print(data_PredIndex_x.head())
print(data_PredIndex_y.head())
```

```python
data_PredIndex_x.to_csv('./data_PredIndex_x.csv', index=False)
data_PredIndex_y.to_csv('./data_PredIndex_y.csv', index=False)
```

## 3) x + i_hat -> y

```python
data_combined=pd.merge(mydata, data_x, on=['년도', '분기', '행정동코드', '표준산업분류코드'], how='left')
data_combined=pd.merge(data_combined, data_pred, on=['년도', '분기', '행정동코드', '표준산업분류코드'], how='left')
data_combined=data_combined.dropna()
data_combined
```

```python
data_combined_x=data_combined.drop('폐업여부', axis=1)
data_combined_y=data_combined[['점포명', '년도', '분기', '행정동코드', '표준산업분류코드', '폐업여부']]

print(data_combined_x.head())
print(data_combined_y.head())
```

```python
data_combined_x.to_csv('./data_combined_x.csv', index=False)
data_combined_y.to_csv('./data_combined_y.csv', index=False)
```

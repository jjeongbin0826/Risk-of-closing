- 공간 데이터 이용해서 행정동 코드 추가함

```python
mydata=pd.read_csv("/content/drive/Shareddrives/서울시 빅데이터(2022-2)/논문/1. 전처리/mydata_dong.csv")
mydata
```

```python
mydata.isnull().sum()

# 행정동코드 매핑 안된게 2개 존재, 처리가 까다로워서 걍 버림.
```

```python
mydata = mydata.dropna(subset=['ADSTRD_CD'])
mydata
```

```python
mydata=mydata.rename(columns={'ADSTRD_CD':'행정동코드'})
mydata=mydata.sort_values(['index']).reset_index(drop=True)
mydata=mydata.drop(columns = ['ADSTRD_NM', 'index'])
mydata['행정동코드']=mydata['행정동코드']//100
mydata
```

```python
mydata['행정동코드']=mydata['행정동코드'].astype('int64')
mydata['개업년월'] = pd.to_numeric(mydata['개업년월'], errors='coerce')
mydata['폐업년월'] = pd.to_numeric(mydata['폐업년월'], errors='coerce')
mydata['표준산업분류코드'] = pd.to_numeric(mydata['표준산업분류코드'], errors='coerce')
mydata.info()
```

```python
mydata = mydata.fillna(0)
mydata = mydata.sort_values(by='표준산업분류코드')
mydata.reset_index(drop=True, inplace=True)
mydata
```

```python
values = []
start_indices = []
change_indices = []

current_value = mydata['표준산업분류코드'].iloc[0]
start_index = 0

for i in range(1, len(mydata['표준산업분류코드'])):
    if mydata['표준산업분류코드'].iloc[i] != current_value:
        values.append(current_value)
        start_indices.append(start_index)
        change_indices.append(i - 1)

        current_value = mydata['표준산업분류코드'].iloc[i]
        start_index = i

values.append(current_value)
start_indices.append(start_index)
change_indices.append(len(mydata['표준산업분류코드']) - 1)

result_df = pd.DataFrame({
    'value': values,
    'start_index': start_indices,
    'change_index': change_indices
})

result_df
```

```python
# 경쟁업체개수 계산

for j in range(len(mydata)):
    row_index = result_df.index[result_df['value'] == mydata.at[j, '표준산업분류코드']].tolist()[0]
    start_index = result_df.at[row_index, 'start_index']
    finish_index = result_df.at[row_index, 'change_index']
    data_mini = mydata.iloc[start_index:finish_index + 1]

    c = len(data_mini[
        (data_mini['개업년월'] < mydata.at[j, '개업년월']) &
        ((data_mini['폐업년월'] > mydata.at[j, '개업년월']) | (data_mini['폐업년월'] == 0)) &
        (data_mini['행정동코드'] == mydata.at[j, '행정동코드'])
    ])

    mydata.at[j, '경쟁업체개수'] = c

mydata
```

```python
filtered_data = mydata[
    (mydata['개업년월'] < mydata.iloc[1]['개업년월']) &
    ((mydata['폐업년월'] > mydata.iloc[1]['폐업년월']) | (mydata['폐업년월'] == 0)) &
    (mydata['행정동코드'] == mydata.iloc[1]['행정동코드']) &
    (mydata['표준산업분류코드'] == mydata.iloc[1]['표준산업분류코드'])
]

filtered_data
```

```python
# 프랜차이즈 비율 계산

unique_hdong_codes = mydata['행정동코드'].unique()

for hdong_code in unique_hdong_codes:
    specific_data = mydata[mydata['행정동코드'] == hdong_code]
    franchise_ratio = specific_data[specific_data['프랜차이즈여부'] == 'Y'].shape[0] / specific_data.shape[0]
    mydata.loc[mydata['행정동코드'] == hdong_code, '프랜차이즈비율'] = franchise_ratio

mydata
```

```python
mydata.to_csv('mydata.csv', index=False)
```

# 데이터 전처리 frame 만들기

```python
mydata=pd.read_csv("/content/drive/Shareddrives/서울시 빅데이터(2022-2)/논문/1. 전처리/mydata.csv")
mydata=mydata[['개업년월', '표준산업분류코드', '행정동코드', '경쟁업체개수', '프랜차이즈비율']]
mydata
```

```python
mydata['년도']=mydata['개업년월']//100
mydata['월']=mydata['개업년월']%100
mydata
```

```python
conditions=[(mydata['월']>=1)&(mydata['월']<=3),
           (mydata['월']>=4)&(mydata['월']<=6),
           (mydata['월']>=7)&(mydata['월']<=9),
           (mydata['월']>=10)&(mydata['월']<=12)]
values=[1,2,3,4]
mydata['분기']=np.select(conditions,values)
mydata
```

```python
mydata=mydata[['년도', '분기', '행정동코드', '표준산업분류코드', '경쟁업체개수', '프랜차이즈비율']]
mydata=mydata.groupby(['년도','분기','행정동코드','표준산업분류코드']).agg({'경쟁업체개수':np.mean,'프랜차이즈비율':np.mean}).reset_index()
mydata
```

```python
original = mydata[['년도', '분기', '행정동코드', '표준산업분류코드']]

original['년도'] = original['년도'] - (original['분기'] == 1)
original['분기'] = original['분기'].apply(lambda x: 4 if x == 1 else x - 1)

original
```

```python
shift_1 = original.copy()

shift_1['년도'] = shift_1['년도'] - (shift_1['분기'] == 1)
shift_1['분기'] = shift_1['분기'].apply(lambda x: 4 if x == 1 else x - 1)

shift_1
```

```python
shift_2 = shift_1.copy()

shift_2['년도'] = shift_2['년도'] - (shift_2['분기'] == 1)
shift_2['분기'] = shift_2['분기'].apply(lambda x: 4 if x == 1 else x - 1)

shift_2
```

```python
frame = pd.concat([shift_1, shift_2], axis=0, ignore_index=True)
frame = frame[(frame['년도']>=2020)&(frame['년도']<=2021)]
frame = frame.drop_duplicates()
frame = frame.sort_values(by=['년도', '분기', '행정동코드', '표준산업분류코드'])
frame = frame.reset_index(drop=True)
frame
```

# frame + 마이데이터

```python
temp=pd.merge(frame, mydata, on=['년도', '분기', '행정동코드', '표준산업분류코드'], how='outer')
temp
```

```python
temp.sort_values(by=['년도', '분기'], inplace=True)
temp['순서'] = temp.groupby(['년도', '분기']).ngroup() + 1

temp
```

```python
temp['프랜차이즈비율'] = temp.groupby(['년도', '분기', '행정동코드'])['프랜차이즈비율'].transform(lambda x: x.fillna(method='ffill').fillna(method='bfill'))
temp.info()
```

```python
for index, row in temp[temp['경쟁업체개수'].isnull()].iterrows():
    candidate_rows = temp[
        (temp['행정동코드'] == row['행정동코드']) &
        (temp['표준산업분류코드'] == row['표준산업분류코드']) &
        ((temp['순서'] == row['순서'] - 1) | (temp['순서'] == row['순서'] + 1))
    ]

    if not candidate_rows.empty:
        avg_value = candidate_rows['경쟁업체개수'].mean()
        temp.at[index, '경쟁업체개수'] = avg_value

temp.info()
```

```python
for index, row in temp[temp['프랜차이즈비율'].isnull()].iterrows():
    candidate_rows = temp[
        (temp['행정동코드'] == row['행정동코드']) &
        ((temp['순서'] == row['순서'] - 1) | (temp['순서'] == row['순서'] + 1))
    ]

    if not candidate_rows.empty:
        avg_value = candidate_rows['프랜차이즈비율'].mean()
        temp.at[index, '프랜차이즈비율'] = avg_value

temp.info()
```

```python
temp['행정동코드']=temp['행정동코드'].astype('int64')
temp=pd.merge(temp,code,on=['행정동코드'], how='left')
temp.info()
```

```python
fill_mean_func = lambda g: g.fillna(g.mean())
temp=temp.groupby(['년도','분기','표준산업분류코드','시군구코드']).apply(fill_mean_func)
temp.info()
```

```python
frame=pd.merge(frame, temp, on=['년도', '분기', '행정동코드', '표준산업분류코드'], how='left')
frame
```

```python
frame=frame.dropna()
frame=frame.drop(['순서', '시군구코드'],axis=1)
frame
```

```python
frame.to_csv('frame.csv', index=False)
```
